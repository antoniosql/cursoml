Las tres series temporales muestran claramente la periodicidad, a menudo llamada estacionalidad en el an√°lisis de las series temporales, en la que un patr√≥n se repite una y otra vez en intervalos de tiempo regulares. Las series temporales de consumo (Consumption), producci√≥n de energ√≠a solar (Solar) y producci√≥n de energ√≠a e√≥lica (Wind) oscilan entre valores altos y bajos en una escala de tiempo anual, que se corresponde con los cambios estacionales del clima a lo largo del a√±o. Sin embargo, la estacionalidad en general no tiene que corresponder con las estaciones meteorol√≥gicas. Por ejemplo, los datos de ventas minoristas a menudo muestran una estacionalidad anual con un aumento de las ventas en noviembre y diciembre, lo que lleva a las vacaciones.

La estacionalidad tambi√©n puede ocurrir en otras escalas de tiempo. El gr√°fico anterior sugiere que puede haber alguna estacionalidad semanal en el consumo de electricidad de Alemania, correspondiente a los d√≠as de semana y fines de semana. Se trazar√° la serie temporal en un solo a√±o para investigar m√°s a fondo.

```
#Comprobar si es estacionaria
from statsmodels.tsa.stattools import adfuller

result = adfuller(opsd_365d_nona['Consumption'].bfill(0))
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Cuanto m√°s negativo es el AD-Fuller, m√°s probable ser√° rechazar la hip√≥tesis nula y m√°s probable que tengamos un dataset estacionario. Para hacerla estacionaria se aplican transformaciones de logaritmo, raiz cuadrada,....

```
import numpy as np
plt.figure(figsize=(12,8))
plt.plot(np.log(opsd_365d_nona['Consumption']))
plt.show()
```

## Eliminar Tendencia

```
from sklearn.linear_model import LinearRegression

X = np.array(range(len(np.log(opsd_365d_nona['Consumption']))))
y = np.log(opsd_365d_nona['Consumption']).ffill(axis=0)

# Linear Trend
reg = LinearRegression().fit(X.reshape(-1,1), y)
pred_lin = reg.predict(X.reshape(-1,1))  

# Logarithmic trend
a_1,b_1 = np.polyfit(np.log(X+1), y, 1)
pred_log = a_1 * np.log(X+1) + b_1

# Exponential trend
a_2,b_2 = np.polyfit(X+1, np.log(y), 1)
pred_exp = np.exp(b_2) + np.exp( (X+1) * a_2)

plt.figure(figsize=(12,8))
plt.plot(np.log(opsd_365d_nona['Consumption']), label="Index")
plt.plot(opsd_365d_nona['Consumption'].index, pred_lin, label="linear trend")
plt.plot(opsd_365d_nona['Consumption'].index, pred_log, label="log trend")
plt.plot(opsd_365d_nona['Consumption'].index, pred_exp, label="exp trend")
plt.legend()
plt.show()
```

```
#Como quitar la tendencia
series = np.log(opsd_365d_nona['Consumption'])-pred_lin
```

```
plt.figure(figsize=(12,8))
plt.plot(series)
plt.show()
```

## Elmininar Estacionalidad

```
# la forma m√°s sencilla para eliminar la tendencia es calcular la primera diferencia, por ejemplo en una estacionalidad anual con medidas mensuales, restarle 12
plt.figure(figsize=(12,8))
plt.plot(series - series.shift(12))
plt.show()
```

```
series_stationary = series - series.shift(12)
result = adfuller(series_stationary.dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

El valor p es cercano a 0, y la estad√≠stica del ADF est√° por debajo del valor cr√≠tico del 1%. Rechazamos la hip√≥tesis nula de que la serie tiene una ra√≠z de unidad y no es estacionaria. ¬°Por lo tanto, la serie es estacionaria!

```
# ARIMA
from statsmodels.tsa.arima_model import ARIMA
size = int(len(series_stationary.dropna()) * 0.75)
train, test = series_stationary.dropna()[0:size], series_stationary.dropna()[size:len(series_stationary.dropna())]
test = test.reset_index()['Consumption']
history = [x for x in train]
predictions = []
```

A continuaci√≥n, para evaluar el rendimiento del modelo, haremos que nuestro modelo prediga el siguiente valor utilizando solo el valor de entrenamiento real. En otras palabras, aplicamos la ventana m√≥vil, y en cada paso, le damos al modelo una observaci√≥n una vez m√°s para predecir desde lugar de usar el √∫ltimo valor predicho. Esta es una manera com√∫n de evaluar mejor el rendimiento de un modelo.

```
for t in range(len(test)):
    model = ARIMA(history, order=(1,1,1))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
```

```
plt.figure(figsize=(12,8))
plt.plot(predictions, label="Prediction")
plt.plot(test, label="Series")
plt.title("ARIMA prediction")
plt.legend()
plt.show()
```

```
from sklearn.metrics import mean_squared_error
mean_squared_error(predictions, test)
```
